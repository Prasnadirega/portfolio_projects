```python
import pandas as pd
import glob
from sqlalchemy import create_engine
import psycopg2
import os
```

```python
files = os.listdir(r'G:\Codecademy\Projects\NewarkAirportBikeRental\data')
pattern = os.path.join(r'G:\Codecademy\Projects\NewarkAirportBikeRental\data', 'JC-*-citibike-tripdata.csv')
bike_files = glob.glob(pattern)
df_list = []
for x in bike_files:
    data = pd.read_csv(x)
    df_list.append(data)
```

### Weather Data

```python
weather = pd.read_csv(r'G:\Codecademy\Projects\NewarkAirportBikeRental\data\newark_airport_2016.csv')
```

```python
weather.head()
```

```python
weather.describe()
```

```python
weather.drop(['PGTM', 'TSUN'], axis=1, inplace=True)
```

```python
weather.describe()
```

```python
weather.head(25)
```

```python
weather.drop(['STATION', 'NAME', 'WDF2', 'WDF5', 'WSF2', 'WSF5'], axis=1, inplace=True)
# I dropped 'STATION' and 'NAME' because they are all the same station.
# As for the fastest winds and their direction, i dont think they are useful because we dont know EXACTLY when those winds happen (i could be wrong though)
```

```python
weather.dtypes
```

```python
weather['DATE'] = pd.to_datetime(weather['DATE'], yearfirst=True)
```

```python
weather.head()
```

```python
weather.columns = ['date', 'avg_wind', 'prcp', 'snow_amount', 'snow_depth', 'avg_temp', 'max_temp', 'min_temp']
```

```python
weather['rain'] = (weather['prcp'] > 0) & (weather['snow_amount'] == 0)
weather['snow'] = weather['snow_amount'] > 0
```

```python
weather.head(15)
```

```python
con = create_engine('postgresql+psycopg2://postgres:postgres@localhost:5432/citibike_data')
```

```python
tripdata.to_sql('citibike_tripdata', con, if_exists='replace', index=False)
```

```python
weather.to_sql('weather_data', con, if_exists='replace', index=False)
```
